### Description
This notebook contains all code for the graphs used to validate the accuracy of the column mapping between the NewFormat and OldFormat MassBay files. From these graphs, it was determined that the column mapping was correct and the units used in the columns were consistent.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dir = '../../data/'
```

#### Unit validation of MassBay OldFormat/NewFormat data

```python
df = pd.read_csv(data_dir + 'MassBay/OldFormat/concat/MassBay_2017_to_2019_CONCAT.csv')
newdf = pd.read_csv(data_dir + 'MassBay/NewFormat/concat/MassBay_2020_to_2022_CONCAT.csv')
```

```python
# Assumed New:Old column mapping
new_to_old = {
 'DISS_OXYGEN (mg/L)' : 'Dissolved Oxygen (Model 43)',
 'TRANS (m-1)' : 'Beam Attenuation',
 'FLU_RAW (ug/L)' : 'Chla Fluor',
 'CONDTVY (mS/cm)' : 'Conductivity',
 'SIGMA_T ()' : 'Sigma-T',
 'SAL (PSU)' : 'Salinity',
 'TEMP (C)': 'Temperature'
}
```

```python
new_indices = [i for i in range(len(newdf))]
old_indices = [i for i in range(len(df))]
```

```python
for new_col in new_to_old:
  old_col = new_to_old[new_col]
  plt.scatter(new_indices, newdf[new_col], label =new_col)
  plt.scatter(old_indices, df[old_col], label = old_col)
  plt.legend()
  plt.show()
```

Determined that units are the same from scatterplots
